# EXPLORING HACKER NEWS POSTS
     In this project, we shall explore the posts in the hacker news website. Both the Ask Hacker News and Show hacker news. The ask Hacker is used by users to ask the community some questions. The show hacker news is used by users to show the community some stuffs they did.

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)

print(hn[0:5])
```

```python
headers = hn[0]
hn=hn[1:]
print(headers)
print(hn[:5])
```

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    title = title.lower()
    
    if title.startswith('ask hn'):
        ask_posts.append(row)
        
    elif title.startswith('show hn'):
        show_posts.append(row)
        
    else:
        other_posts.append(row)
        
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
    
```

```python
#total_ask_comments = len(ask_posts)
total_ask_comments = 0
for row in ask_posts:
    number_of_comments = int(row[4])
    total_ask_comments += number_of_comments
    avg_ask_comments = total_ask_comments / len(ask_posts)
    print(avg_ask_comments)
    
    
total_show_comments = 0
for row in show_posts:
    total_show_comments += int(row[4])
    avg_show_comments = sum(total_show_comments) / len(show_posts)
print(avg_show_comments)    
```
